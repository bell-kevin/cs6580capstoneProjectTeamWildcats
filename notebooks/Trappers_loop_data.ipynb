```python
import pandas as pd

def load_trappers_loop_data(start_year=2015, end_year=2024):
    all_frames = []
    
    for year in range(start_year, end_year + 1):
        # Construct the path dynamically
        path = f"../data/raw/trappersLoopCounts/Trappers_Loop_{year}.csv"
        
        try:
            # Read the file
            temp_df = pd.read_csv(path)
            all_frames.append(temp_df)
            print(f"Successfully loaded: {year}")
        except FileNotFoundError:
            print(f"Warning: File for {year} not found at {path}")
            
    # Combine all individual dataframes into one
    if all_frames:
        return pd.concat(all_frames, ignore_index=True)
    else:
        return pd.DataFrame() 

# Load data
df = load_trappers_loop_data()
```

```python
# Drop unnecessary columns
df.drop(columns=["STATION", "ROUTE", "MP"], inplace=True)
```

```python
# Unpivot dataframe
id_vars = ['DATE', 'LANE']
df_long = df.melt(id_vars=id_vars, var_name='Hour', value_name='Count')
```

```python
# Clean the Hour column
df_long['Hour'] = df_long['Hour'].str.extract('(\\d{4})')

# Combine DATE and Hour into a single datetime object
df_long['Date_Hour'] = pd.to_datetime(
    df_long['DATE'] + ' ' + df_long['Hour'], 
    format='%m/%d/%Y %H%M'
)
```

```python
# Drop last columns, order by date hour and reset index
df_clean = df_long[['Date_Hour', 'LANE', 'Count']]
df_clean = df_clean.sort_values(by=['Date_Hour', 'LANE'], ascending=True)
df_clean = df_clean.reset_index(drop=True)
```
