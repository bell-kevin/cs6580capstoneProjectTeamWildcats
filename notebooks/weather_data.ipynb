## Load Weather Data into DFs

```python
import pandas as pd

df_2016 = pd.read_excel("../data/raw/weatherData/SBBWK_2016.xlsx")
df_2017 = pd.read_excel("../data/raw/weatherData/SBBWK_2017.xlsx")
df_2018 = pd.read_excel("../data/raw/weatherData/SBBWK_2018.xlsx")
df_2019 = pd.read_excel("../data/raw/weatherData/SBBWK_2019.xlsx")
df_2020 = pd.read_excel("../data/raw/weatherData/SBBWK_2020.xlsx")
df_2021 = pd.read_excel("../data/raw/weatherData/SBBWK_2021.xlsx")
df_2022 = pd.read_excel("../data/raw/weatherData/SBBWK_2022.xlsx")
df_2023 = pd.read_excel("../data/raw/weatherData/SBBWK_2023.xlsx")
df_2024 = pd.read_excel("../data/raw/weatherData/SBBWK_2024.xlsx")
df_2025 = pd.read_excel("../data/raw/weatherData/SBBWK_2025.xlsx")
df_2026 = pd.read_excel("../data/raw/weatherData/SBBWK_2026.xlsx")

df = pd.concat([
    df_2016,  
    df_2017,
    df_2018,
    df_2019,
    df_2020,
    df_2021,
    df_2022,
    df_2023,
    df_2024,
    df_2025,
    df_2026
])

df.to_csv("../data/interim/weather_data.csv", index=False)

```
